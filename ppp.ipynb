### aggreate function

```python
def data_aggreate(df_new,groupby_columns):
    grouped_data = df_new.groupby(groupby_columns)
    size = grouped_data.size().reset_index(name='counts')
    rural_size = grouped_data['RuralUrbanIndicator'].apply(lambda x: x[x.str.contains('R')].count()).reset_index(name='rural_counts')['rural_counts']
    hub_size = grouped_data['HubzoneIndicator'].apply(lambda x: x[x.str.contains('Y')].count()).reset_index(name='hub_counts')['hub_counts']
    lmi_size = grouped_data['LMIIndicator'].apply(lambda x: x[x.str.contains('Y',na = False)].count()).reset_index(name='lmi_counts')['lmi_counts']
    data_counts = size
    data_counts['rural_size'] =rural_size
    data_counts['hub_size'] =hub_size
    data_counts['lmi_size'] =lmi_size
    data_counts
    jobs_mean = grouped_data['JobsReported'].mean().reset_index(name = 'jobs_mean')['jobs_mean']
    jobs_sum = grouped_data['JobsReported'].sum().reset_index(name = 'jobs_sum')['jobs_sum']
    init_approval_mean = grouped_data['InitialApprovalAmount'].mean().reset_index(name = 'init_approval_mean')['init_approval_mean']
    init_approval_sum = grouped_data['InitialApprovalAmount'].sum().reset_index(name = 'init_approval_sum')['init_approval_sum']
    data_ratio = data_counts
    data_ratio['rural_ratio'] = data_ratio['rural_size']/data_ratio['counts']
    data_ratio['hub_ratio'] = data_ratio['hub_size']/data_ratio['counts']
    data_ratio['lmi_ratio'] = data_ratio['lmi_size']/data_ratio['counts']
    data_ratio['jobs_mean'] = jobs_mean
    data_ratio['jobs_sum'] = jobs_sum
    data_ratio['init_approval_mean'] = init_approval_mean
    data_ratio['init_approval_sum'] = init_approval_sum
    data_ratio = data_ratio.sort_values(by = ['approved_year','approved_month'])
    return data_ratio
```

### start analysis

```python
groupby_columns = ['approved_month','approved_year','BorrowerState','NAICSCode']
data1 = data_aggreate(df_new1,groupby_columns)
data1
```
